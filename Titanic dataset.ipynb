**checking for missing values**

```python
missing_train=train.isnull().sum().sort_values(ascending=False)
missing_train_pct=(train.isnull().mean()*100).sort_values(ascending=False)
pd.concat([missing_train,missing_train_pct],axis = 1,keys = ["missing_counts","missing_percentage"])
```

**The dataset has significant missingness in Cabin (~77%) and Age (~20%). The Embarked column has less than 1% missing values, while the rest are complete. This will influence how we clean or impute the data before modeling.”**



```python
missing_test=test.isnull().sum().sort_values(ascending=False)
missing_percentage=(test.isnull().mean()*100).sort_values(ascending=False)
pd.concat([missing_test,missing_percentage],axis=1,keys=['missing_counts','missing_percentage'])
```

**The dataset has significant missingness in Cabin (~79%) and Age (~21%). The Embarked column has less than 1% missing values, while the rest are complete. This will influence how we clean or impute the data before modeling.**



**TREATING MISSING VALUES**

```python
train["Age"]=train["Age"].fillna(train['Age'].median())
test["Age"]=test["Age"].fillna(test['Age'].median())
train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0])
test['Embarked']=test['Embarked'].fillna(test['Embarked'].mode()[0])
```

```python
test['Fare']=test['Fare'].fillna(test['Fare'].median())
```

```python
train.isnull().sum().sort_values(ascending=False)
```

```python
test.isnull().sum().sort_values(ascending=False)
```

```python
train['HasCabin']=train['Cabin'].notnull().astype(int)
test['HasCabin']=test['Cabin'].notnull().astype(int)
```

**EDA VISUALIZATIONS**

```python
#Compare survivors (1) vs non-survivors (0).
sns.countplot(x='Survived', data=train)
plt.title("Survival Counts")
plt.show()
```

```python
#See how class (1st/2nd/3rd) affects survival.
sns.countplot(x='Pclass',hue='Survived',data=train)
plt.title("Survival by class")
plt.show()
```

```python
sns.countplot(x='Sex',hue='Survived',data=train)
plt.title('Survival by gender')
plt.show()
```

**females has higher survival rate**

```python
sns.histplot(train['Age'],bins=30,kde=True)
plt.title('Age Distribution')
plt.show()
```

**The age distribution shows that most passengers were young adults and middle-aged, with a peak around 20 to 30 years. 
There are fewer children and elderly passengers, and the distribution is slightly right-skewed, indicating more younger passengers onboard.”**



```python
#survival vs age

sns.boxplot(x='Survived',y='Age',data=train)
plt.title("Age vs Survival")
plt.show()
```

```python
numeric_cols = train.select_dtypes(include=['number'])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**The correlation heatmap shows the relationships between numerical features. Fare is positively correlated with survival,
indicating passengers who paid more had higher chances of surviving.
Pclass is negatively correlated with survival, meaning first-class passengers had better survival rates.
Age shows little correlation, suggesting age alone was less influential.
Other features like SibSp and Parch show moderate correlation, reflecting family connections onboard.**

```python
sns.barplot(x='HasCabin',y='Survived',data=train)
plt.title("Hascabin vs Survived")
plt.show()
```

```python
sns.scatterplot(x="Age",y="Fare",hue='Survived',data=train,palette='coolwarm')
plt.title("age vs fare coloured by survival")
plt.show()
```

**Passengers who paid higher fares (mostly 1st class) tend to have higher survival rates (many points in the upper right with survival=1).**

**Younger passengers with low fares (mostly 3rd class) have many who did not survive.**

**There is a cluster of non-survivors paying low fare across all ages.**

```python
sns.pairplot(train, vars=['Age', 'Fare', 'SibSp', 'Parch'], hue='Survived', palette='coolwarm')
plt.suptitle('Pairplot of Numeric Features by Survival', y=1.02)
plt.show()
```

**Survivors tend to have paid higher fares on average.**

**Age distribution shows some survivors are younger, but there are survivors across all ages.**

**Number of siblings/spouses (sibsp) and parents/children (parch) don’t show a strong difference by survival in pairwise plots but may still be factors.**

**Fare and survival show positive correlation.**
